```python
import requests

BASE_URL = "http://localhost:3002"

def test_create_user():
    payload = {
        "name": "mewband12",
        "time_off_balance": 10,
        "title": "Consultant",
        "address": "123 Main St, Bangkok"
    }
    response = requests.post(f"{BASE_URL}/create-user", json=payload)
    print("Create User:", response.json())

def test_user_profile():
    payload = {"name": "mewband12"}
    response = requests.post(f"{BASE_URL}/user_profile_details", json=payload)
    print("User Profile:", response.json())

def test_time_off_balance():
    payload = {"name": "mewband12"}
    response = requests.post(f"{BASE_URL}/time-off-balance", json=payload)
    print("Time Off Balance:", response.json())

def test_request_time_off():
    payload = {
        "name": "mewband12",
        "from_date": "2025-06-25",
        "to_date": "2025-06-28"
    }
    response = requests.post(f"{BASE_URL}/request-time-off", json=payload)
    print("Request Time Off:", response.json())

def test_update_title():
    payload = {
        "name": "mewband12",
        "new_title": "Senior Consultant"
    }
    response = requests.put(f"{BASE_URL}/update-title", json=payload)
    print("Update Title:", response.json())

def test_update_address():
    payload = {
        "name": "mewband12",
        "new_address": "456 Innovation St, Singapore"
    }
    response = requests.put(f"{BASE_URL}/update-address", json=payload)
    print("Update Address:", response.json())

def test_get_emails():
    payload = {"name": "mewband12"}
    response = requests.post(f"{BASE_URL}/get_emails", json=payload)
    print("Emails:", response.json())

def test_get_card_transactions():
    payload = {"name": "mewband12"}
    response = requests.post(f"{BASE_URL}/get_corporate_card_transactions", json=payload)
    print("Corporate Card Transactions:", response.json())

def test_create_expense_claim():
    payload = {
        "name": "mewband12",
        "sustainable_travel_incentive": True,
        "expense_report": "Report_001"
    }
    response = requests.post(f"{BASE_URL}/create-expense-claim", json=payload)
    print("Create Expense Claim:", response.json())

def test_export_users():
    response = requests.get(f"{BASE_URL}/export-users/")
    print("Export Users:", response.status_code)
    with open("exported_users.xlsx", "wb") as f:
        f.write(response.content)
    print("Downloaded 'exported_users.xlsx'")

def test_clear_users():
    response = requests.delete(f"{BASE_URL}/clear-users/")
    print("Clear Users:", response.json())

if __name__ == "__main__":
    test_create_user()
    test_user_profile()
    test_time_off_balance()
    test_request_time_off()
    test_update_title()
    test_update_address()
    test_get_emails()
    test_get_card_transactions()
    test_create_expense_claim()
    test_export_users()
    test_clear_users()
```
